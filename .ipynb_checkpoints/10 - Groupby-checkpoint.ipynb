```python
import pandas as pd
# Cria um DataFrame
data = {'Classe':['Junior','Junior','Pleno','Pleno','Senior','Senior'],
       'Nome': ['Jorge','Carlos','Roberta','Patricia','Bruno','Vera'],
       'Venda':[200,120,340,124,243,350]}

df = pd.DataFrame(data)
df
```

```python
# agrupando por classe e fznd a soma,
group = df.groupby("Classe") # primeiro ele agrupa a func, dps passamos a operacao
group.sum()
```

```python
# media
group.mean()
```

```python
# mostrando os menores de cada Classe
df.groupby("Classe").min()
```

```python
# os maiores de cada Classe
df.groupby("Classe").max()
```

```python
#df2 = df # assim qnd eu altero os valores da coluna Venda do df2, ele altera do df1 tbm
df2 = df.copy() # df2 sera Independente, qnd alterar a coluna do df2, nn fara alteração no df
df
```

```python
df2["Venda"] = [150, 432, 190, 230, 410, 155]
df2
```

```python
# ele junta os DFs,
df3 = pd.concat([df, df2])
df3
```

```python
# df3 = pd.concat([df, df2])
# juntando as Informações pelos indices
df3.groupby(["Classe", "Nome"]).sum()
```
